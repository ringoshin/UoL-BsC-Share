#This is re-factored from MYK's basic.py into a working Colab notebook that will display the outcome of each action in a notebook environment.

```python
import os
import gymnasium as gym
from gymnasium.wrappers import RecordVideo # Wrapper to record videos of the environment
import ale_py # Dependency for Atari environments in Gymnasium
from gymnasium.wrappers import AtariPreprocessing, FrameStackObservation # Wrappers for Atari game preprocessing

from IPython.display import HTML # For displaying HTML content in Jupyter/Colab notebooks (e.g., videos)
import base64
import glob
import io
```

```python
def show_video():
    """Finds and displays the first .mp4 video in the 'video/' directory."""
    mp4list = glob.glob('video/*.mp4')
    if len(mp4list) > 0:
        mp4 = mp4list[0]
        video = io.open(mp4, 'r+b').read()
        encoded = base64.b64encode(video)
        return HTML(data='''<video alt="test" autoplay
                loop controls style="height: 400px;">
                <source src="data:video/mp4;base64,{0}" type="video/mp4" />
             </video>'''.format(encoded.decode('ascii')))
    else:
        return "Video not found"
```

```python
# Create and wrap the environment
env = gym.make("BreakoutNoFrameskip-v4", render_mode="rgb_array")
env = RecordVideo(env, './video', episode_trigger=lambda x: True)
```

```python
state, info = env.reset()
for _ in range(1000):
    action = env.action_space.sample()
    state, reward, terminated, truncated, info = env.step(action)

    if terminated or truncated:
        break

env.close()
```

```python
show_video()
```
